```python
#Ejercicio 17 - Samuel López
import math

class Calculos:
  @staticmethod
  def cal_longitud_circunferencia(radio):
    return 2 * math.pi * radio

  @staticmethod
  def cal_area_circulo(radio):
    return math.pi * math.pow(radio,2)

radio = 3.0

longitud_circunferencia = Calculos.cal_longitud_circunferencia(radio)
area_circulo = Calculos.cal_area_circulo(radio)

print("Longitud circunferencia: ", longitud_circunferencia)
print("Área circulo: ", area_circulo)
```
